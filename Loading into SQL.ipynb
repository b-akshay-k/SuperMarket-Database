```python
import pandas as pd
import sqlalchemy as sal

# Database connection parameters
username = 'root'
password = 'Akshay123'
database_name = 'mart'
host = '127.0.0.1'
port = '3306'

# MySQL connection string
connection_string = f'mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database_name}'

# Create the SQLAlchemy engine
engine = sal.create_engine(connection_string)

# Check the connection
try:
    with engine.connect() as connection:
        print("Connection successful!")
except Exception as e:
    print(f"Connection error: {e}")

# Load the transactions data into a DataFrame
redemptions = pd.read_csv("C:\\Users\\admin\\OneDrive\\Documents\\Important Files\\Customer Segmentation\\Redemption.csv")

# Ensure column names match the MySQL table schema
redemptions.columns = ['RedemptionID', 'CustomerID', 'UsedCouponID', 'DiscountAmount', 'RedemptionDate']

# Load the customers table to get valid CustomerID values
query = "SELECT CustomerID FROM customers"
customers = pd.read_sql(query, con=engine)

# Filter the transactions DataFrame to include only valid CustomerID values
valid_customers = customers['CustomerID'].tolist()
filtered_redemptions = redemptions[redemptions['CustomerID'].isin(valid_customers)]

# Write the DataFrame to the MySQL table
try:
    filtered_redemptions.to_sql('redemptions', con=engine, index=False, if_exists='append')
    print("Data successfully uploaded!")
except Exception as e:
    print(f"Data upload error: {e}")
```
